```python
import pandas
import shutil
```

```python
class PreProcessor:
    def execute(self):
        raise NotImplementedError
```

```python
class HH101PreProcessor(PreProcessor):
    def __init__(self):
        self.raw_fp = '../resources/raw/hh101/hh101/rawdata.txt'
        self.processed_fp = '../resources/processed/hh101.pre.dat'

    def execute(self):
        shutil.copy2(self.raw_fp, self.processed_fp)

        with open(self.processed_fp, 'r') as file:
            content = file.read()

        with open(self.processed_fp, 'w') as file:
            file.write(' '.join(['date', 'time', 'thing', 'data']) + '\n')
            file.write(content)

        processed_hh101 = pandas.read_csv(self.processed_fp, sep=' ')
        
        print(processed_hh101)
```

```python
if __name__ == '__main__':
    hh101_pp = HH101PreProcessor()
    hh101_pp.execute()
```
